```python
import shap
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

USERNAME = "shreya"
BASE_PATH = f"/Users/{USERNAME}/Desktop/ARIA"
MODEL_PATH = f"{BASE_PATH}/models"

# Load all three models
model_hc = joblib.load(f"{MODEL_PATH}/healthcare_model.pkl")
model_ft = joblib.load(f"{MODEL_PATH}/fintech_model.pkl")
model_saas = joblib.load(f"{MODEL_PATH}/saas_model.pkl")

features_hc = joblib.load(f"{MODEL_PATH}/healthcare_features.pkl")
features_ft = joblib.load(f"{MODEL_PATH}/fintech_features.pkl")
features_saas = joblib.load(f"{MODEL_PATH}/saas_features.pkl")

print("✅ All models loaded")
```

```python
# Recreate healthcare data
np.random.seed(42)
n = 25000

df_hc = pd.DataFrame({
    'age': np.random.randint(18, 90, n),
    'time_in_hospital': np.random.randint(1, 14, n),
    'n_lab_procedures': np.random.randint(1, 100, n),
    'n_procedures': np.random.randint(0, 6, n),
    'n_medications': np.random.randint(1, 30, n),
    'n_outpatient': np.random.randint(0, 10, n),
    'n_inpatient': np.random.randint(0, 5, n),
    'n_emergency': np.random.randint(0, 5, n),
    'num_diagnoses': np.random.randint(1, 16, n),
    'has_diabetes': np.random.randint(0, 2, n),
    'has_hypertension': np.random.randint(0, 2, n),
    'has_heart_disease': np.random.randint(0, 2, n),
    'insulin_prescribed': np.random.randint(0, 2, n),
    'discharge_type': np.random.randint(1, 10, n),
    'admission_source': np.random.randint(1, 8, n),
})

X_hc = df_hc[features_hc]

# SHAP explainer
explainer_hc = shap.TreeExplainer(model_hc)
shap_values_hc = explainer_hc.shap_values(X_hc[:500])

# Plot feature importance
plt.figure()
shap.summary_plot(shap_values_hc, X_hc[:500], plot_type="bar", show=False)
plt.title("Healthcare — Top Risk Factors")
plt.tight_layout()
plt.savefig(f"{BASE_PATH}/docs/shap_healthcare.png")
plt.show()
print("✅ Healthcare SHAP done")
```

```python
df_ft = pd.read_csv(f"{BASE_PATH}/data/fintech_clean.csv")
X_ft = df_ft[features_ft]

explainer_ft = shap.TreeExplainer(model_ft)
shap_values_ft = explainer_ft.shap_values(X_ft[:500])

plt.figure()
shap.summary_plot(shap_values_ft, X_ft[:500], plot_type="bar", show=False)
plt.title("Fintech — Top Fraud Risk Factors")
plt.tight_layout()
plt.savefig(f"{BASE_PATH}/docs/shap_fintech.png")
plt.show()
print("✅ Fintech SHAP done")
```

```python
from sklearn.preprocessing import LabelEncoder

df_saas = pd.read_csv(f"{BASE_PATH}/data/saas_clean.csv")
df_saas['TotalCharges'] = pd.to_numeric(df_saas['TotalCharges'], errors='coerce')
df_saas['TotalCharges'].fillna(df_saas['TotalCharges'].median(), inplace=True)

cat_cols = df_saas.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    df_saas[col] = le.fit_transform(df_saas[col].astype(str))

X_saas = df_saas[features_saas]

explainer_saas = shap.TreeExplainer(model_saas)
shap_values_saas = explainer_saas.shap_values(X_saas[:500])

plt.figure()
shap.summary_plot(shap_values_saas, X_saas[:500], plot_type="bar", show=False)
plt.title("SaaS — Top Churn Risk Factors")
plt.tight_layout()
plt.savefig(f"{BASE_PATH}/docs/shap_saas.png")
plt.show()
print("✅ SaaS SHAP done")
```

```python
def get_risk_explanation(model, explainer, features, input_data):
    """
    Given one patient/user/transaction — return risk score + top 3 reasons
    """
    df_input = pd.DataFrame([input_data])[features]
    
    # Risk score
    prob = model.predict_proba(df_input)[0][1]
    risk_score = round(prob * 100, 1)
    
    # SHAP explanation
    shap_vals = explainer.shap_values(df_input)[0]
    
    # Top 3 factors
    factors = sorted(
        zip(features, shap_vals),
        key=lambda x: abs(x[1]),
        reverse=True
    )[:3]
    
    return {
        "risk_score": risk_score,
        "risk_level": "HIGH" if risk_score >= 70 else "MEDIUM" if risk_score >= 40 else "LOW",
        "top_factors": [
            {"feature": f, "impact": round(v * 100, 1)}
            for f, v in factors
        ]
    }

# Test with a sample patient
sample_patient = {
    'age': 72,
    'time_in_hospital': 8,
    'n_lab_procedures': 60,
    'n_procedures': 3,
    'n_medications': 20,
    'n_outpatient': 0,
    'n_inpatient': 3,
    'n_emergency': 2,
    'num_diagnoses': 9,
    'has_diabetes': 1,
    'has_hypertension': 1,
    'has_heart_disease': 1,
    'insulin_prescribed': 1,
    'discharge_type': 3,
    'admission_source': 7
}

result = get_risk_explanation(model_hc, explainer_hc, features_hc, sample_patient)
print(f"\nRisk Score: {result['risk_score']}/100")
print(f"Risk Level: {result['risk_level']}")
print("\nTop Risk Factors:")
for f in result['top_factors']:
    print(f"  {f['feature']}: {f['impact']:+.1f} points")
```
